### Problem 10 — Keyword vs positional
Predict all outputs.
```python

    return x, y, z

```

````python
# Prediction: (1,5,7) (1,5,9), (1,9,7)

def g(x, y=5, z=7):
    return x, y, z
print(g(1))
print(g(1, z=9))
print(g(1, 9))

# Explanation: In the first call, only x is given, so y and z use their default values of 5 and 7. In the second call, x is 1 and z is set to 9 by name, so y stays at its default value of 5. In the third call it is 1 9 7 because 9 is assigned to y

````

### Problem 13 — Default parameter trap
Predict the outputs.


````python
# Prediction: (1) (1,2)

def add_item(item, bucket=[]):
    bucket.append(item)
    return bucket
print(add_item(1))
print(add_item(2))

# Explanation: Each time the function is called without providing a bucket, it uses the same list. The first call adds 1 to the list. The second call adds 2 to the same list, so the result becomes 1,2

````
